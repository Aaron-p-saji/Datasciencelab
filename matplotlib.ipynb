```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([2,4,6,8,10,12])
y=np.array([3,5,7,9,6,14])
x1=np.array([6,3,3,7,8,2])
y1=np.array([3,5,7,9,6,14])
plt.plot(x,y,label='1st')
plt.plot(x1,y1,label='2nd')
plt.legend(loc='upper right')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,10*np.pi,1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,5*np.pi,1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,z)
plt.xlabel("values")
plt.ylabel("sin and cos")
plt.title("plot sin and cos")
plt.legend(['sin(x)','cos(x)'],loc='upper right')
plt.show()
```
